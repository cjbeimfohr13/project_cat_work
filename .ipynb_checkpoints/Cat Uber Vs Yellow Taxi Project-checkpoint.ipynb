```python
r = requests.get(url)
yellow_x = r.json()
Yellow_Taxi_df = pd.DataFrame(yellow_x)
Yellow_Taxi_df.head()

Yellow_Taxi_df.dropna(how="any")
Yellow_Taxi_df["pickup_datetime"]=pd.to_datetime(Yellow_Taxi_df.pickup_datetime)
Yellow_Taxi_df['new_date'] = [d.date() for d in Yellow_Taxi_df['pickup_datetime']]
Yellow_Taxi_df['new_time'] = [d.time() for d in Yellow_Taxi_df['pickup_datetime']]

Yellow_Taxi_df["Month"]= Yellow_Taxi_df.pickup_datetime.dt.month

# yellowtaxi_lat=Yellow_Taxi_df["pickup_latitude"].to_list()
# yellowtaxi_lon=Yellow_Taxi_df["pickup_longitude"].to_list()

Yellow_Taxi_df.dtypes


Yellow_Taxi_df2 = Yellow_Taxi_df.replace(0, np.nan)
Yellow_Taxi_df3 = Yellow_Taxi_df2.dropna(how='all', axis=0)


Tax_lat=round(Yellow_Taxi_df3["pickup_latitude"].apply(pd.to_numeric),4) 
Tax_lon=round(Yellow_Taxi_df3["pickup_longitude"].apply(pd.to_numeric),4)

Tax_lat_df=pd.DataFrame(Tax_lat)
Tax_lat_df

Tax_lon_df=pd.DataFrame(Tax_lon)


TLA=Tax_lat_df["pickup_latitude"].to_list()
TLO=Tax_lon_df["pickup_longitude"].to_list()

TLA_TLO=pd.DataFrame({"Taxi_Latitude":TLA, "Taxi_Longitude":TLO, "new_date":Yellow_Taxi_df3['new_date'],"Time":Yellow_Taxi_df3['new_time'], "Taxi_Pickup":Yellow_Taxi_df3["pickup_datetime"], "Taxi_Month":Yellow_Taxi_df3["Month"]})

# Clean_Yellow_df=Yellow_Taxi_df[["pickup_datetime","new_date", "new_time","Month"]]
TLA_TLO

bool_series = TLA_TLO["Taxi_Pickup"].between("4/01/2014", "09/30/2014", inclusive = True)
Taxi_Data_final=TLA_TLO[bool_series]
Taxi_Data_final
```

```python
plt.scatter(Taxi_Data_final.Taxi_Latitude,Taxi_Data_final.Taxi_Longitude,marker="o", color="b",zorder=5)
plt.ylabel('longitude')
plt.xlabel('latitude')
plt.title('Location of Taxi Pickups (popular locations)')

plt.savefig("Scatter_Taxi")
```

```python
Taxi_Data_final.Taxi_Month.value_counts().sort_index().plot()
plt.ylabel("Number of rides")
plt.xlabel("Months")
plt.title("Number of Taxi Rides per Month")
plt.savefig("Yellow_TaxiRides_per_month")
```

```python
csv_path="C:/Users/cjbei/Desktop/project_cat_work/uber-raw-data-apr14.csv"
csv_path1="C:/Users/cjbei/Desktop/project_cat_work/uber-raw-data-may14.csv"
csv_path2="C:/Users/cjbei/Desktop/project_cat_work/uber-raw-data-jun14.csv"
csv_path3="C:/Users/cjbei/Desktop/project_cat_work/uber-raw-data-jul14.csv"
csv_path4="C:/Users/cjbei/Desktop/project_cat_work/uber-raw-data-aug14.csv"
csv_path5="C:/Users/cjbei/Desktop/project_cat_work/uber-raw-data-sep14.csv"
```

```python
april=pd.read_csv(csv_path)
uber_april=pd.DataFrame(april)

may=pd.read_csv(csv_path1)
uber_may=pd.DataFrame(may)

june=pd.read_csv(csv_path2)
uber_june=pd.DataFrame(june)

july=pd.read_csv(csv_path3)
uber_july=pd.DataFrame(july)

aug=pd.read_csv(csv_path4)
uber_aug=pd.DataFrame(aug)

sept=pd.read_csv(csv_path5)
uber_sept=pd.DataFrame(sept)
```

```python
uber_merge=pd.concat([uber_april,uber_may,uber_june,uber_july,uber_aug,uber_sept])
uber_merge_df=uber_merge.rename(columns={"Date/Time":"DateTime"})
uber_merge_df
```

```python
uber_merge_df.dtypes
```

```python
uber_merge_df["DateTime"]=pd.to_datetime(uber_merge_df.DateTime)
uber_merge_df

uber_merge_df['new_date'] = [d.date() for d in uber_merge_df['DateTime']]
uber_merge_df['new_time'] = [d.time() for d in uber_merge_df['DateTime']]

uber_merge_df

# uber_merge.Date/Time.str.slice(-7,-3).astype(int).head()
```

```python
uber_merge_df["Month"]= uber_merge_df.DateTime.dt.month
uber_merge_df
```

```python
uber_merge_df.Month.value_counts().sort_index().plot()
plt.ylabel("Number of Rides")
plt.xlabel("Months")
plt.title("Uber Number of Rides Per Month")

plt.savefig("Uberridesper_month")
```

```python
plt.scatter(uber_merge_df.Lat,uber_merge_df.Lon,marker="o", color="r", zorder=5)
plt.ylabel('longitude')
plt.xlabel('latitude')
plt.title('Location of Uber Pickups (popular locations)')

plt.savefig("Scatter_Uber")
```
